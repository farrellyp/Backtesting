## Section 2: Strategy Development

**Requirement:** Select, explain and implement a trading strategy for a currency pair

<p style='text-align: justify;'> When developing a strategy it is important to remove as much bias as possible. By fitting models to historical data and testing in the same sample set can result in overfitting. An alternative is to develop strategies using historical data and test in a simulated environment. In this case I have randomly selected a Simple Moving Average Crossover Strategy using 200 and 20 day moving averages and thus it is not subject to such bias. </p>

<p style='text-align: justify;'> The Simple Moving Average Crossover Strategy is a Momentum or Trend-Following Strategy. It entails the use of two moving average technical indicators. These are defined using a rolling average over a stated period. When the short-term moving average indicator crosses the long-term moving average it is regarded as bullish or a buy signal. Similarly when it returns below the long-term moving average this is regarded as bearish or a signal to sell the asset. One challenge with moving averages is that they require enough historical data to meet their definition and compute the moving average accurately. For this reason, the strategy was implemented and tested only affter the first 200 hours of data. </p>

<p style='text-align: justify;'> To implement this in python, a rolling window of appropriate length for each moving average was defined, from which an average was calculated for each time point. To generate the buy and sell signals a Signal column was initiated in which the value was 0 if the 20-day moving average was below the 200-day and 1 if the opposite was true. Then a Position column was created to calculate the difference between the current row and the previous one. If this difference is non-zero a change in position will occur. The results of this can be seen in the image below with the zoomed in subplot highlighting how the moving average crossovers create a change in position. </p>

## Section 1: Data Collection

````python
import yfinance as yf
import pandas as pd
import numpy as np
from scipy import stats
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, zoomed_inset_axes, mark_inset
````

````python
#Load in EURUSD data from yfinance

def get_forex_data(symbol, start_date, end_date, interval='1h'):
    """
    Download forex data using yfinance and add estimated bid/ask prices.
    
    :param symbol: Forex pair symbol (e.g., 'EURUSD=X')
    :param start_date: Start date for data download
    :param end_date: End date for data download
    :param interval: Data interval (default is '1h' for hourly)
    :return: DataFrame with OHLC data and estimated bid/ask prices
    """
    # Download data
    data = yf.download(symbol, start=start_date, end=end_date, interval=interval)
    
    # Estimate spread (Simple Approach)
    spread_pips = 2  # Assume 2 pips spread, adjust as needed
    spread = spread_pips * 0.0001  # Convert pips to price for major pairs
    
    # Calculate estimated bid and ask prices
    data['Bid'] = data['Close'] - spread / 2
    data['Ask'] = data['Close'] + spread / 2
    
    return data

# Implementation
if __name__ == "__main__":
    symbol = "EURUSD=X"
    end_date = datetime.now()
    start_date = end_date - timedelta(days=365)  # Get last 365 days of data
    
    forex_data = get_forex_data(symbol, start_date, end_date)
    
    print(forex_data.head())
    print(f"\nTotal rows: {len(forex_data)}")
    
````

## Section 2 & 3: Strategy Development, Visulaisations & Backtest

````python
# Calculate 20-day and 200-day moving averages
forex_data['SMA20'] = forex_data['Close'].rolling(window=20).mean()
forex_data['SMA200'] = forex_data['Close'].rolling(window=200).mean()

#forex_data.head()
forex_data.tail()
````

````python
def generate_signals(forex_data):
    # Make a copy of the DataFrame to avoid SettingWithCopyWarning
    df = forex_data.copy()
    
    # Initialize the Signal column with zeros
    df['Signal'] = 0
    
    # Generate signals based on moving average crossover
    df.loc[df.index[200:], 'Signal'] = np.where(df['SMA20'][200:] > df['SMA200'][200:], 1, 0)
    
    # Calculate the Position column
    df['Position'] = df['Signal'].diff()
    
    return df

# Apply the function to dataframe
forex_data = generate_signals(forex_data)

# Display the first few rows
print(forex_data.head())

# Display some statistics about the signals
print("\nSignal Statistics:")
print(forex_data['Signal'].value_counts())
print("\nPosition Statistics:")
print(forex_data['Position'].value_counts())
````

````python
# Creating a visualisation with zoomed in subplot

def plot_ma_crossover_strategy(forex_data, zoom_start, zoom_end):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 12), gridspec_kw={'height_ratios': [2, 1]})
    
    # Plot the full timeline
    ax1.plot(forex_data['Close'], label='EUR/USD Close Price')
    ax1.plot(forex_data['SMA20'], label='20-day SMA')
    ax1.plot(forex_data['SMA200'], label='200-day SMA')
    ax1.plot(forex_data[forex_data['Position'] == 1].index, 
             forex_data['SMA20'][forex_data['Position'] == 1], 
             '^', markersize=10, color='g', label='Buy signal')
    ax1.plot(forex_data[forex_data['Position'] == -1].index, 
             forex_data['SMA20'][forex_data['Position'] == -1], 
             'v', markersize=10, color='r', label='Sell signal')
    ax1.set_title('EUR/USD Simple Moving Average Crossover Strategy')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Price')
    ax1.legend()
    
    # Plot the zoomed-in view
    zoom_data = forex_data.loc[zoom_start:zoom_end]
    ax2.plot(zoom_data['Close'], label='EUR/USD Close Price')
    ax2.plot(zoom_data['SMA20'], label='20-day SMA')
    ax2.plot(zoom_data['SMA200'], label='200-day SMA')
    ax2.plot(zoom_data[zoom_data['Position'] == 1].index, 
             zoom_data['SMA20'][zoom_data['Position'] == 1], 
             '^', markersize=10, color='g', label='Buy signal')
    ax2.plot(zoom_data[zoom_data['Position'] == -1].index, 
             zoom_data['SMA20'][zoom_data['Position'] == -1], 
             'v', markersize=10, color='r', label='Sell signal')
    ax2.set_title('Zoomed View')
    ax2.set_xlabel('Date')
    ax2.set_ylabel('Price')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

# Example usage
zoom_start = '2024-06-01'  # Adjust these dates to focus on a specific period
zoom_end = '2024-09-06'
plot_ma_crossover_strategy(forex_data, zoom_start, zoom_end)
````

````python
def calculate_strategy_returns(forex_data):
    # Make a copy of the DataFrame to avoid SettingWithCopyWarning
    df = forex_data.copy()
    
    # Calculate returns using Ask price for buys and Bid price for sells
    df['BuyReturns'] = df['Ask'].pct_change()
    df['SellReturns'] = df['Bid'].pct_change()
    
    # Calculate strategy returns
    df['StrategyReturns'] = np.where(
        df['Position'].shift(1) == 1,  # Buy signal
        df['BuyReturns'],
        np.where(
            df['Position'].shift(1) == -1,  # Sell signal
            -df['SellReturns'],  # Negative return for short positions
            0  # No position
        )
    )
    
    # Calculate cumulative returns for the strategy
    df['CumulativeStrategyReturns'] = (1 + df['StrategyReturns']).cumprod() - 1
    
    # Calculate buy-and-hold returns using Close price
    df['Returns'] = df['Close'].pct_change()
    df['CumulativeBuyHoldReturns'] = (1 + df['Returns']).cumprod() - 1
    
    return df

# Apply the function to the dataframe
forex_data = calculate_strategy_returns(forex_data)

# Display the first few rows
print(forex_data[['Close', 'Ask', 'Bid', 'Position', 'StrategyReturns', 'CumulativeStrategyReturns']].head())

# Display summary statistics
print("\nStrategy Returns Summary:")
print(forex_data['StrategyReturns'].describe())

print("\nCumulative Strategy Returns at the end of the period:")
print(forex_data['CumulativeStrategyReturns'].iloc[-1])

print("\nCumulative Buy & Hold Returns at the end of the period:")
print(forex_data['CumulativeBuyHoldReturns'].iloc[-1])
````

````python
# Visualize cumulative returns
plt.figure(figsize=(12, 8))
plt.plot(forex_data['CumulativeStrategyReturns'], label='Strategy Returns')
plt.plot(forex_data['CumulativeBuyHoldReturns'], label='Buy & Hold Returns')
plt.title(f'EUR/USD Strategy vs. Buy & Hold Cumulative Returns')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.legend()
plt.show()
````

````python
#Risk Measures

def calculate_risk_measures(returns, frequency='hourly'):
    """
    Calculate risk measures for a given series of returns.
    
    :param returns: pandas Series of strategy returns
    :param frequency: string, either 'hourly' or 'daily'
    :return: dictionary of risk measures
    """
    
    # Annualization factor
    if frequency == 'hourly':
        annual_factor = 252 * 24  # For 252 trading days and 24 trading hours per day
    elif frequency == 'daily':
        annual_factor = 252
    else:
        raise ValueError("Frequency must be either 'hourly' or 'daily'")
    
    risk_measures = {}
    
    # 1. Sharpe Ratio
    risk_measures['Sharpe Ratio'] = np.sqrt(annual_factor) * returns.mean() / returns.std()
    
    #2. Max Drawdown
    cum_returns = (1 + returns).cumprod()
    peak = cum_returns.expanding(min_periods=1).max()
    drawdown = (cum_returns / peak) - 1
    risk_measures['Maximum Drawdown'] = drawdown.min()
    
    return risk_measures

#Apply to dataframe
strategy_returns = forex_data['StrategyReturns']
risk_measures = calculate_risk_measures(strategy_returns, frequency='hourly')

# Print the risk measures
for measure, value in risk_measures.items():
    print(f"{measure}: {value:.4f}")

# Compare with buy-and-hold strategy
buy_hold_returns = forex_data['Returns']
buy_hold_risk_measures = calculate_risk_measures(buy_hold_returns, frequency='hourly')
print("\nBuy & Hold Strategy Risk Measures:")
for measure, value in buy_hold_risk_measures.items():
    print(f"{measure}: {value:.4f}")
````
